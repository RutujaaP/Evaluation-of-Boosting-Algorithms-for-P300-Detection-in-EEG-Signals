```python
from hyperopt import hp,fmin,tpe,STATUS_OK,Trials
```

```python
space = {
    'learning_rate':     hp.choice('learning_rate',     [0.05, 0.31, 0.07]),
    'max_depth':         hp.choice('max_depth',         np.arange(5, 16, 1, dtype=int)),
     'colsample_bylevel': hp.choice('colsample_bylevel', [0.3, 0.8, 0.1]),
    'n_estimators':      hp.choice('n_estimators',[1000,1200,1500]),
    'eval_metric':       'BalancedAccuracy',
       }

```

```python
 def objective(space):
    model = CatBoostClassifier(task_type='GPU',
                               learning_rate= space['learning_rate'],
                               max_depth =space['max_depth'],
                               n_estimators=space['n_estimators'])
                                # colsample_bylevel=space['colsample_bylevel'])
                                
    accuracy = cross_val_score(model, X_train, Y_train, cv = 5).mean()

    # We aim to maximize accuracy, therefore we return it as a negative value
    return {'loss': -accuracy, 'status': STATUS_OK }

from sklearn.model_selection import cross_val_score
trials = Trials() #trials is responsible for minimizing the function value
best = fmin(fn= objective,
            space= space,
            algo= tpe.suggest,
            max_evals = 100,
            trials= trials)
```
